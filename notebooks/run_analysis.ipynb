```python
import matplotlib.pyplot as plt

plt.bar(df["Ligand"], df["Affinity"])
plt.xlabel("Ligand")
plt.ylabel("Binding Affinity (kcal/mol)")
plt.title("Docking Scores")
plt.xticks(rotation=45)
plt.show()
```

```python
print(f"Total ligands screened: {len(df)}")

best = df.sort_values("Affinity").iloc[0]

print(f"Best ligand: {best['Ligand']}")
print(f"Binding affinity: {best['Affinity']} kcal/mol")
```
